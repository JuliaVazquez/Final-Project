## Import libraries and Dataframe

```python
import pandas as pd
```

```python
style = pd.read_csv('scraping_style.csv')
style.shape
```

```python
style
```

```python
def cleaning_wine(df):
    print(df.shape)
    display(df[df['wine_id'].duplicated()==True])
    cleaned_df = df.drop_duplicates()
    print(cleaned_df.shape)
    display(cleaned_df[cleaned_df['wine_id'].duplicated()==True])
    display(cleaned_df.isna().sum())
    return cleaned_df
```

```python
df = cleaning_wine(style)
```

## Drop rows that have not information (in the scraping fase they where empty)

```python
rows_drop = df[df['wine_id'] == df['body']].index
```

```python
cleaned = df.drop(rows_drop)
print(df.shape[0] - len(rows_drop))
cleaned.shape
```

## Check NaNs

### Fill NaNs with 0 because I think it is the missing value

```python
update = cleaned.fillna(0)
```

## Round values 

```python
def rounded(df,cols):
    for col in cols:
        df[col] = round(df[col],2)
    return df
```

```python
cols = ['acidity_2','fizziness','intensity','sweetness','tannin']
update = rounded(update,cols)
update
```
