## Import libraries and dataframe

```python
import pandas as pd
import re
```

```python
df = pd.read_csv('scraping_wines.csv')
```

```python
df
```

## Droping unnamed column

```python
def drop(df):
    df = df.drop(['Unnamed: 0'],axis=1)
    return df
```

```python
wine = drop(df)
wine
```

## Checking for duplicates

```python
wine['wine_name'].value_counts()
```

```python
wine[wine['wine_name']=='Bodegas Lleiroso LL Reserva 2015']
```

## Drop duplicated rows

```python
cleaned_wine = wine.drop_duplicates()
cleaned_wine
```

## Check duplicated wine_ids

```python
cleaned_wine['wine_name'].value_counts()
```

```python
duplicated_id = cleaned_wine[cleaned_wine['wine_id'].duplicated()==True]['wine_id'].values
```

```python
cleaned_wine[cleaned_wine['wine_id'].isin(duplicated_id)].sort_values(by='wine_id')
```

```python
cleaned_wine.shape
```

## Drop wine_ids with less num_ratings

```python
drop_indexes = [20868,9058,20869,20877,20911,20876,20905,20920]
```

```python
update_wine = cleaned_wine.drop(drop_indexes)
update_wine.shape
```

## Check and drop NaNs

```python
update_wine.isna().sum()
```

```python
update_wine[update_wine['year'].isna()]['wine_name'].values
```

```python
update_wine = update_wine.dropna()
```

## Take out year inside wine_name

```python
update_wine['wine_name'].value_counts()
```

```python
new_names = []

for i in update_wine['wine_name']:
    new_names.append(re.sub(r'\d{4}', '', i))
```

```python
update_wine['wine_name'] = new_names
update_wine
```

## Save csv

```python
update_wine.to_csv('update_wine.csv',index=False)
```
