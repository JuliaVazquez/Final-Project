```python
def getting_df(pages, cols_names):
    
    results_total = []
    
    for j in range(len(pages)):
        results = [(
            t["vintage"]["id"],
            t["vintage"]["name"],
            t['vintage']['wine']['winery']['name'],
            t['vintage']['wine']['type_id'],
            t['vintage']['wine']['region']['name'],
            t['vintage']['wine']['region']['country']['seo_name'],
            t['vintage']['year'],
            t['vintage']['statistics']['wine_ratings_count'],
            t['vintage']['statistics']['wine_ratings_average']
        )
    for t in pages[j].json()['explore_vintage']['matches']
    ]
        results_total.append(results)
    
    
    df1 = pd.DataFrame(results_total[0],columns=cols_names)
    
    
    for i in range(1,len(results_total)):
        df = pd.DataFrame(results_total[i],columns=cols_names)
        df1 = pd.concat([df1,df],axis=0)
    
    return df1
```

```python
def getting_style(pages, cols_names):
    results_total = []
    
    for j in range(len(pages)):
        print(j)
        for i in range(len(pages[j].json()['explore_vintage']['matches'])):
            if pages[j].json()['explore_vintage']['matches'][i]['vintage']['wine']['style']!=None:
                if len(pages[j].json()['explore_vintage']['matches'][i]['vintage']['wine']['style'])>7:
                    t = pages[j].json()['explore_vintage']['matches'][i]["vintage"]
                    s = pages[j].json()['explore_vintage']['matches'][i]["vintage"]['wine']['taste']['structure']
                    results = [(
                        t["id"],
                        t['wine']['style']['varietal_name'],
                        t['wine']['style']['body'],
                        t['wine']['style']['acidity'],
                        t['wine']['style']['acidity_description'],
                        s['acidity'],
                        s['fizziness'],
                        s['intensity'],
                        s['sweetness'],
                        s['tannin']
                    )]
                    results_total.append(results)
                else:
                    h = pages[j].json()['explore_vintage']['matches'][i]["vintage"]
                    l = pages[j].json()['explore_vintage']['matches'][i]['vintage']['wine']['style']['id']
                    results = [(
                        h["id"],
                        h['wine']['style']['varietal_name'],
                        l,
                        l,
                        l,
                        l,
                        l,
                        l,
                        l,
                        l
                    )]
                    results_total.append(results)
            else:
                r = pages[j].json()['explore_vintage']['matches'][i]["vintage"]["id"]
                results = [(
                    r,
                    r,
                    r,
                    r,
                    r,
                    r,
                    r,
                    r,
                    r,
                    r
                )]
                results_total.append(results)
                
    df1 = pd.DataFrame(results_total[0],columns=cols_names)
    
    
    for k in range(1,len(results_total)):
        df = pd.DataFrame(results_total[k],columns=cols_names)
        df1 = pd.concat([df1,df],axis=0)
    
    return df1
```
