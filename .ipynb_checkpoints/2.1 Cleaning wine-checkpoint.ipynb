## Import libraries and dataframe

```python
import pandas as pd
import re
```

```python
wine = pd.read_csv('scraping_wines.csv')
```

```python
wine
```

## Checking for duplicates

```python
wine['wine_name'].value_counts()
```

```python
wine[wine['wine_name']=='Bodegas Lleiroso LL Reserva 2015']
```

## Drop duplicated rows

```python
cleaned_wine = wine.drop_duplicates()
cleaned_wine
```

```python
cleaned_wine[cleaned_wine['wine_name'] == 'Bodegas Lleiroso LL Reserva 2015']
```

## Check duplicated wine_ids

```python
cleaned_wine['wine_name'].value_counts()
```

```python
cleaned_wine[cleaned_wine['wine_id'].duplicated()==True]
```

```python
cleaned_wine.shape
```

## Check and drop NaNs

```python
cleaned_wine.isna().sum()
```

```python
cleaned_wine[cleaned_wine['year'].isna()]['wine_name'].values
```

```python
update_wine = cleaned_wine.dropna()
update_wine.shape
```

## Take out year inside wine_name

```python
update_wine['wine_name'].value_counts()
```

```python
new_names = []

for i in update_wine['wine_name']:
    new_names.append(re.sub(r'\d{4}', '', i))
```

```python
update_wine['wine_name'] = new_names
update_wine
```

## Save csv

```python
update_wine.to_csv('update_wine.csv',index=False)
```
