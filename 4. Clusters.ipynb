# Import libraries

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
%matplotlib inline

from sklearn import cluster, datasets
from sklearn.preprocessing import StandardScaler, MinMaxScaler, OneHotEncoder
from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans
```

# Import data

```python
pd.set_option('display.max_columns', None)

df = pd.read_csv('data_to_model.csv')
df
```

## Split, scale and encode

```python
df.select_dtypes(object)
```

```python
df.select_dtypes(np.number)
```

```python
df['style_id'].value_counts()
```

```python
df['totals'] = df['grapes_1_count']+df['grapes_2_count']+df['grapes_3_count']
df['percent_1'] = round(df['grapes_1_count']/df['totals'],2)
df['percent_2'] = round(df['grapes_2_count']/df['totals'],2)
df['percent_3'] = round(df['grapes_3_count']/df['totals'],2)
df
```

```python
def get_clusters(df):

    #split numericals - categoricals
    X_cat = df.select_dtypes(object)
    X_num = df[['body','acidity_1','acidity_2','fizziness','intensity','sweetness','tannin','num_ratings','rate']]

        
    #fit and transform scaling numericals MinMax
    X_num_mm = MinMaxScaler().fit_transform(X_num) 
    # convert to dataframe again
    X_num_mm = pd.DataFrame(X_num_mm,columns=X_num.columns).reset_index(drop=True) 

    
    
    #fit and transform scaling numericals Standard
    X_num_st = StandardScaler().fit_transform(X_num) 
    # convert to dataframe again
    X_num_st = pd.DataFrame(X_num_st,columns=X_num.columns).reset_index(drop=True) 
    
    
    
    #fit and transform encoding categoricals
    X_cat = X_cat.astype(str)
    onehot = OneHotEncoder(handle_unknown='error', drop='first').fit(X_cat) 

    cols = onehot.get_feature_names_out(input_features=X_cat.columns)
    X_cat_enc = onehot.transform(X_cat).toarray()    
    # convert to dataframe again
    X_cat_enc = pd.DataFrame(X_cat_enc,columns=cols).reset_index(drop=True)

    
    return X_num_mm, X_num_st, X_cat_enc
```

```python
X_num_mm, X_num_st, X_cat_enc = get_clusters(df)
```

```python
X_num_mm
```

## Check how many clusters

#### with minmax scaled data

```python
%%time

K = range(2, 35)
inertia = []

for k in K:
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(X_num_mm)
    inertia.append(kmeans.inertia_)


plt.figure(figsize=(16,8))
plt.plot(K, inertia, 'bx-')
plt.xlabel('k')
plt.ylabel('inertia')
plt.xticks(np.arange(min(K), max(K)+1, 1.0))
plt.title('Elbow Method showing the optimal k')
```

```python
%%time

K = range(2, 35)
silhouette = []

for k in K:
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(X_num_mm)
    silhouette.append(silhouette_score(X_num_mm, kmeans.predict(X_num_mm)))


plt.figure(figsize=(16,8))
plt.plot(K, silhouette, 'bx-')
plt.xlabel('k')
plt.ylabel('silhouette score')
plt.xticks(np.arange(min(K), max(K)+1, 1.0))
plt.title('Silhouette score showing the optimal k')
```

```python
%%time

K = range(30, 45)
silhouette = []

for k in K:
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(X_num_mm)
    silhouette.append(silhouette_score(X_num_mm, kmeans.predict(X_num_mm)))


plt.figure(figsize=(16,8))
plt.plot(K, silhouette, 'bx-')
plt.xlabel('k')
plt.ylabel('silhouette score')
plt.xticks(np.arange(min(K), max(K)+1, 1.0))
plt.title('Silhouette score showing the optimal k')
```

## Making the clusters

```python
kmeans = KMeans(n_clusters=39)
kmeans.fit(X_num_mm)
```

```python
# Assigning the clusters:
clusters = kmeans.predict(X_num_mm)

# Check the size of the clusters
pd.Series(clusters).value_counts().sort_index()
```

```python
# visualizing clusters with PCA
from sklearn.decomposition import PCA
```

```python
pca = PCA()
pca.fit(X_num_mm)
```

```python
pca.explained_variance_ratio_.round(2)
```

```python
X_num_mm_pca = pca.transform(X_num_mm)
```

```python
options, charts = plt.subplots(1, 2, figsize=(12, 6))

charts[0].scatter(X_num_mm_pca[:,0],X_num_mm_pca[:,1], c = clusters)
charts[1].scatter(X_num_mm_pca[:,0],X_num_mm_pca[:,2], c = clusters)
plt.show()
```

```python
options, charts = plt.subplots(1, 2, figsize=(12, 6))

charts[0].scatter(X_num_mm_pca[:,1],X_num_mm_pca[:,2], c = clusters)
charts[1].scatter(X_num_mm_pca[:,2],X_num_mm_pca[:,3], c = clusters)
plt.show()
```
